```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('winequality-red.csv',delimiter=';')
tf = df[1400:1600]
df = df[0:1399]
import df
```

```python
#==========Part 1===========#

import cvxpy as cp

'''Training'''
df_np = df.to_numpy(dtype='float')
n = 11

Y = [df_np[:,11]]
X = df_np[:,0:11]
beta = cp.Variable(1)
w = cp.Variable((11,1))

cost = (cp.sum_squares(X@w + beta - Y))
res = cp.Problem(cp.Minimize(cost)).solve()


'''Testing'''
tf_np = tf.to_numpy(dtype='float')
loss = 0

for i in range(0, (np.size(tf_np, 0))):
    param = tf_np[i][:-1]
    expected = tf_np[i][-1]

    res = param @ w.value + beta.value
    loss += abs(res[0]-expected)

print(loss/199)
```

```python
##Using Least squares
#Training model
train_np = df_np

Y = [train_np[:,11]]
X = train_np[:,0:11]

X = np.c_[np.ones(len(X)),X]

inv_X = np.linalg.pinv(X)
theta = np.dot(inv_X,np.transpose(Y))



#Testing Model
test_np = tf_np
loss = 0
X_test = test_np[:,0:11]
X_test = np.c_[np.ones(len(X_test)),X_test]
Y_test= [test_np[:,11]]
Y_prediction = np.dot(X_test,theta)
loss = 0

for i in range(199):
    loss += abs(Y_prediction[i] - Y_test[0][i])
loss/=199
print(loss[0])
```

```python
#==========Part 2===========#


import cvxpy as cp

'''Training'''
df_np = df.to_numpy(dtype='float')
n = 11

Y = [df_np[:,11]]
X = df_np[:,0:11]
beta = cp.Variable(1)
w = cp.Variable((11,1))

cost = cp.sum(cp.huber(X@w + beta - Y, 1))
res = cp.Problem(cp.Minimize(cost)).solve()


'''Testing'''
tf_np = tf.to_numpy(dtype='float')
loss = 0

for i in range(0, (np.size(tf_np, 0))):
    param = tf_np[i][:-1]
    expected = tf_np[i][-1]

    res = param @ w.value + beta.value
    loss += abs(res[0]-expected)

print(loss/199)
```

```python
#==========Part 3===========#


import cvxpy as cp

'''Training'''
df_np = df.to_numpy(dtype='float')
n = 11

Y = [df_np[:,11]]
X = df_np[:,0:11]
beta = cp.Variable(1)
w = cp.Variable((11,1))

cost = cp.abs((X*w) + beta - Y)
cost = cp.maximum(cost-0.5, 0)
res = cp.Problem(cp.Minimize(cp.sum(cost))).solve(solver=cp.ECOS)

'''Testing'''
tf_np = tf.to_numpy(dtype='float')
loss = 0

for i in range(0, (np.size(tf_np, 0))):
    param = tf_np[i][:-1]
    expected = tf_np[i][-1]

    res = param @ w.value + beta.value
    loss += abs(res[0]-expected)
print(loss/199)
```
